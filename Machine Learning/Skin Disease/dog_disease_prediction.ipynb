```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.models import load_model
import numpy as np
```

```python
sample_path = '/content/drive/MyDrive/model/'
```

```python
from PIL import Image

model = load_model('/content/drive/MyDrive/model/disease_dogs.h5')

image_paths = ['/content/drive/MyDrive/model/redness_1.jpg',
               '/content/drive/MyDrive/model/keratosis_1.jpg',
               '/content/drive/MyDrive/model/lesion_2.jpeg']

class_labels = ['Circular Alopecia',
                'Conjunctival Injection or Redness',
                'Keratosis',
                'Nasal Discharge',
                'Ocular Discharge',
                'Skin Lesions']
images = []
for path in image_paths:
    img = Image.open(path)
    img = img.resize((224, 224)) 
    img = np.array(img) / 255.0 
    images.append(img)

images = np.array(images)

predictions = model.predict(images)

for i in range(len(image_paths)):
    img_path = image_paths[i]
    prediction = predictions[i]
    class_index = np.argmax(prediction)
    predicted_class = class_labels[class_index]
    confidence = prediction[class_index] * 100

    # Menampilkan hasil prediksi
    print('img', img_path, ':', predicted_class, 'with', confidence, '% confidence')
```
